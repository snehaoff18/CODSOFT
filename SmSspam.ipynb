```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split #seperating train and test data
from sklearn.feature_extraction.text import CountVectorizer#coverts text document to matrix
from sklearn.naive_bayes import MultinomialNB #it classifies with discrete features
#naive bayes is a part of generative learning algorithm
```

```python
path="/content/drive/MyDrive/spam.csv"
df=pd.read_csv(path, encoding='latin-1') #reading csv files
```

```python
import numpy as np
import nltk #natural language processing
```

```python
df=df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1) #it removes specific rows and columns
```

```python
df=df.rename(columns={"v1":"label","v2":"text"})
df.head()#display top rows of data frame
```

```python
df.describe() #returns description of data in data frame
```

```python
df['spam']=df['label'].apply(lambda x: 1 if x=='spam' else 0) #lamda takes any no of arguments
df['spam'].value_counts() #returns object count of unique values
```

```python
df.head() # shows top rows of dataframe
```

```python
spam=df[df.spam==1] #spam==1
ham=df[df.spam==0] # ham=0
```

```python
n_sample=ham.sample(n=4000)#it returns particular length list of items
n_sample.head() #shows top rows of data set
```

```python
n_dataset=pd.concat([n_sample,spam],axis=0) #it concats 2 type of data set as single data frame
n_dataset['spam'].value_counts()#counts the value
```

```python
x_train,x_test,y_train,y_test=train_test_split(n_dataset.text,n_dataset.spam,test_size=0.30) #splitting the data
```

```python
cv=CountVectorizer()
x_train_count=cv.fit_transform(x_train.values)# used for scaling the data and also to learn scaling parameters
```

```python
model=MultinomialNB()
model.fit(x_train_count,y_train) #to minimize loss of function by adjusting the model parameters
```

```python
#prediction test (ham)
target_ham_value=0
email_ham=[input("enter a text :")]
email_ham_count=cv.transform(email_ham)
prediction=model.predict(email_ham_count) #predicts the output
if np.array_equal(prediction,np.array([target_ham_value])):#if two elements have same array and same shape
   print("its not a spam")
else:
  print("this is spam")
```

```python
#prediction test (Spam)
#spam count
target_ham_value=0
email_spam=[input("enter a text :")]
email_spam_count=cv.transform(email_spam)
prediction=model.predict(email_spam_count)#predicts output
if np.array_equal(prediction,np.array([target_ham_value])):
   print("its not a spam")
else:
  print("this is spam")
```

```python
#accuracy score
x_test_count=cv.transform(x_test)
model.score(x_test_count,y_test)
```
